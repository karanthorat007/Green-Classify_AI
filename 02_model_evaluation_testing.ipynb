ğŸ”¹ Cell 1: Imports

```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
```

ğŸ”¹ Cell 2: Load Pretrained Model

```python
MODEL_PATH = "vegetable_classifier_model.h5"
model = load_model(MODEL_PATH)

model.summary()
```

ğŸ”¹ Cell 3: Test Data Generator


```python
test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(
    "dataset/test",
    target_size=(224, 224),
    batch_size=32,
    class_mode="categorical",
    shuffle=False
)
```

ğŸ”¹ Cell 4: Evaluate Model Accuracy

```python
loss, accuracy = model.evaluate(test_generator)
print(f"Test Accuracy: {accuracy*100:.2f}%")
```

ğŸ”¹ Cell 5: Predictions


```python
predictions = model.predict(test_generator)
y_pred = np.argmax(predictions, axis=1)
y_true = test_generator.classes
```

ğŸ”¹ Cell 6: Classification Report

```python
from sklearn.metrics import classification_report

class_labels = list(test_generator.class_indices.keys())

print(classification_report(y_true, y_pred, target_names=class_labels))
```

ğŸ”¹ Cell 7: Test Single Image

```python
from tensorflow.keras.preprocessing import image

def predict_image(img_path):
    img = image.load_img(img_path, target_size=(224,224))
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    
    prediction = model.predict(img_array)
    class_index = np.argmax(prediction)
    
    return class_labels[class_index]

# Example
print(predict_image("dataset/test/Tomato/1001.jpg"))
```
